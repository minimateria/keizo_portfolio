```python
import requests
from bs4 import BeautifulSoup

# 取得したいページのURLを指定
url = 'https://beauty.hotpepper.jp/slnH000183903/?cstt=299'

# ページのコンテンツを取得
response = requests.get(url)
content = response.content

# BeautifulSoupオブジェクトを作成
# BeautifulSoupはHTMLやXMLからデータを抽出するためのライブラリで、contentとパーサーの種類を引数に指定します。
soup = BeautifulSoup(content, 'html.parser')

# 目的のテキストを取得
# soup.select_oneメソッドは、CSSセレクタに一致する最初の要素を取得します。
# この例では、html body要素内の9番目のscript要素を取得しています。
# nth-of-type(9)は、9番目の子要素（9番目のscript要素）を意味します。
target_text = soup.select_one('html body script:nth-of-type(9)').string

# 結果の表示
print(target_text)
```

```python
import requests

def get_postal_code(address):
    api_url = "http://zipcoda.net/api"

    # APIに渡すパラメータを設定
    params = {
        "address": address
    }

    try:
        response = requests.get(api_url, params=params)
        data = response.json()

        if data["items"]:
            postal_code = data["items"][0]["zipcode"]
            return postal_code
        else:
            print("該当する郵便番号が見つかりませんでした。")
            return None

    except requests.exceptions.RequestException as e:
        print("エラーが発生しました:", e)
        return None

if __name__ == "__main__":
    address = input("address? > :")
    postal_code = get_postal_code(address)
    if postal_code:
        print(f"{address} の郵便番号は {postal_code} です。")
```

```python
#店舗詳細データ取得完成版

import requests
from bs4 import BeautifulSoup
import csv

def get_target_text(url):
    # ページのコンテンツを取得
    response = requests.get(url)
    content = response.content

    # BeautifulSoupオブジェクトを作成
    # BeautifulSoupはHTMLやXMLからデータを抽出するためのライブラリで、contentとパーサーの種類を引数に指定します。
    soup = BeautifulSoup(content, 'html.parser')

    # 目的のテキストを取得
    target_text = soup.select_one('html body script:nth-of-type(9)').string

    # @typeが"BreadcrumbList"の場合は10番目のscript要素を取得
    if target_text and '"@type": "BreadcrumbList"' in target_text:
        target_text = soup.select_one('html body script:nth-of-type(10)').string

    return target_text

# CSVファイルからURLと店名を読み込む
urls_and_store_names = []
with open('/content/drive/MyDrive/salon_data.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        urls_and_store_names.append({'url': row['リンク'], 'store_name': row['店名']})

# データを取得してcsvファイルに書き込む
with open('/content/drive/MyDrive/output_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['店名', '目的のテキスト']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for data in urls_and_store_names:
        url = data['url']
        store_name = data['store_name']

        target_text = get_target_text(url)
        print(target_text)

        # 取得したデータをcsvファイルに書き込む
        writer.writerow({'店名': store_name, '目的のテキスト': target_text})
```

```python
#店舗詳細データ取得excelフォーマット完成版

import requests
from bs4 import BeautifulSoup
import csv
from tqdm import tqdm
import json
import pandas as pd

def get_target_text(url):
    # ページのコンテンツを取得
    response = requests.get(url)
    content = response.content

    # BeautifulSoupオブジェクトを作成
    soup = BeautifulSoup(content, 'html.parser')

    # script要素をすべて取得
    script_tags = soup.find_all('script')

    # JSONデータを含むscript要素を探索
    for script_tag in script_tags:
        script_text = script_tag.string
        if script_text and '"@type": "BeautySalon"' in script_text:
            # JSONデータをパースして目的のテキストを取得
            data = json.loads(script_text)
            return data

    return None

# CSVファイルからURLを読み込む
urls = []
with open('/content/drive/MyDrive/salon_data.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        urls.append(row['リンク'])

# データを取得してDataFrameに格納
data_list = []
for url in tqdm(urls[:1000], desc="処理中", ascii=True):
    target_data = get_target_text(url)
    if target_data:
        # JSONデータを成型してリストに追加
        flattened_data = {}
        for key, value in target_data.items():
            if isinstance(value, dict):
                for sub_key, sub_value in value.items():
                    flattened_data[f"{key}:{sub_key}"] = sub_value
            else:
                flattened_data[key] = value

        # URLからcodeを抽出
        code = url.split("https://beauty.hotpepper.jp/", 1)[-1].split("/", 1)[0]
        flattened_data['code'] = code

        data_list.append(flattened_data)

# DataFrameを作成
df = pd.DataFrame(data_list)

# Excelファイルとしてエクスポート
df.to_excel('/content/drive/MyDrive/salon_list.xlsx', index=False)
```
